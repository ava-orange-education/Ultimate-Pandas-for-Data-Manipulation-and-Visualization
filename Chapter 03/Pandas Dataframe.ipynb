```python
# Create a DataFrame
data = {
   'fruit': ['apple', 'banana', 'cherry', 'date'],
   'color': ['red', 'yellow', 'red', 'brown'],
   'weight': [120, 150, 45, 60]
}

df = pd.DataFrame(data)
df
```

```python

# Adding a single column
new_column = [10, 20, 30, 40]
df['price'] = new_column
df
```

```python
# Adding multiple columns
new_columns = pd.DataFrame({'taste': ['sweet', 'sweet', 'sour', 'sweet'],
                            'season': ['winter', 'summer', 'spring', 'autumn']})

df = pd.concat([df, new_columns], axis=1)
df
```

```python
df.drop('price', axis=1, inplace=True)

df
```

```python
df.drop(['season', 'taste'], axis=1, inplace=True)
df
```

```python
##############Reordering Columns##
```

```python
# Create a sample DataFrame
data = {'fruit': ['apple', 'banana', 'cherry'],
        'color': ['red', 'yellow', 'red'],
        'weight': [120, 150, 45]}
df = pd.DataFrame(data)
df
```

```python
# Reordering columns using indexing
df = df[['color', 'weight', 'fruit']]

df
```

```python
# Reordering columns using reindex()
new_order = ['color', 'fruit', 'weight']
df = df.reindex(columns=new_order)
df
```

```python
data = {'fruit': ['apple', 'banana', 'cherry'],
        'color': ['red', 'yellow', 'red'],
        'weight': [120, 150, 45]}
df = pd.DataFrame(data)
df
```

```python

# Reordering columns using the loc[] indexer
df = df.loc[:, ['color', 'fruit', 'weight']]
df
```

```python
# Reordering columns using the iloc[] indexer
df = df.iloc[:, [1, 0, 2]]

df
```

```python
data = {
    'Name': ['John', 'Alice', 'Bob', 'Emily', 'Michael'],
    'Age': [30, 25, 35, 27, 32],
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance'],
    'Salary': [50000, 60000, 55000, 45000, 70000]
}

df = pd.DataFrame(data)
df
```

```python
filtered_df = df[df['Age'] > 30]
filtered_df
```

```python
filtered_df = df[(df['Age'] > 30) & (df['Department'] == 'IT')]
filtered_df
```

```python
filtered_df = df[df['Department'].str.startswith('H')]
filtered_df
```

```python
filtered_df = df[df['Department'].isin(['HR', 'Finance'])]
filtered_df
```

```python
filtered_df = df[df['Salary'].isnull()]
filtered_df
```

```python
filtered_df = df[df['Salary'].notnull()]
filtered_df
```

```python
###########Arithematic Operations##############
```

```python
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
}

df = pd.DataFrame(data)
df
```

```python
# Adding a scalar to the DataFrame
df_add_scalar = df + 10
df_add_scalar
```

```python
# Adding two DataFrames element-wise
df_add_df = df + df
df_add_df
```

```python
# Subtracting a scalar from the DataFrame
df_subtract_scalar = df - 5
df_subtract_scalar
```

```python
# Subtracting two DataFrames element-wise
df_subtract_df = df - df
df_subtract_df
```

```python
# Multiplying the DataFrame by a scalar
df_multiply_scalar = df * 2
df_multiply_scalar
```

```python
# Element-wise multiplication of two DataFrames
df_multiply_df = df * df
df_multiply_df
```

```python
# Dividing the DataFrame by a scalar
df_divide_scalar = df / 3
df_divide_scalar
```

```python
# Element-wise division of two DataFrames
df_divide_df = df / df
df_divide_df
```
